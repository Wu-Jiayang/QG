```python
import numpy as np
import pandas as pd

data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
df = pd.DataFrame(data, index=labels)                                           # Create df with labels as index
df
```

```python
information = df.info()                                                         # Show df basic information and its data
information
```

```python
columns = df[['animal', 'age']]                                                 # Select only the 'animal' and 'age' columns
columns
```

```python
missingvalue = df[df['age'].isnull()]                                           # Select the row with missing value,
missingvalue
```

```python
sort = df.sort_values(by=['age', 'visits'], ascending=[False, True])            # Sort DF in descending age and ascending visit order
sort
```

```python
df = df.replace('snake', 'python')                                              # In the 'animal' column, replace 'Snake' with 'Python'
df
```
